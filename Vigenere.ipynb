```python
def vigenere(message, key):
    alphabet = ""
    for i in range(32,255):
        if i >= 127 and i <= 160:
            continue
        else:
            alphabet = alphabet + chr(i)
    output = ''
    keycounter = 0
    keylen = len(key) - 1
    for i in message:
        if keycounter > keylen:
            keycounter = 0
        if i in alphabet:
            mesval = alphabet.index(i)
            keyval = alphabet.index(key[keycounter])
            newval = mesval + keyval
            if newval >= len(alphabet):
                newval = newval - len(alphabet)
            newchar = alphabet[newval]
            output += newchar
            keycounter += 1         
        else:
            output += i
    return output

def descrambler(message, key):
    alphabet = ""
    for i in range(32,255):
        if i >= 127 and i <= 160:
            continue
        else:
            alphabet = alphabet + chr(i)
    output = ''
    keycounter = 0
    keylen = len(key) - 1
    for i in message:
        if keycounter > keylen:
            keycounter = 0
        if i in alphabet:
            mesval = alphabet.index(i)
            keyval = alphabet.index(key[keycounter])
            newval = mesval - keyval
            if newval >= len(alphabet):
                newval = newval - len(alphabet)
            newchar = alphabet[newval]
            output += newchar
            keycounter += 1         
        else:
            output += i
    return output

def gcd(a,b):
    if a > b:
        x = a
        y = b
    else:
        x = b
        y = a
    if x%y == 0:
        return(y)
    else:
        while x%y != 0: 
            c = int(x/y)
            d = x%y
            x = y
            y = d
    return(d)

def LCM(lister):
    counter = 0 
    x = lister
    lcmlist = []
    for i in range(0, len(lister) - 1):
        numer = lister[i] * lister[i + 1]
        denom = gcd(lister[i], lister[i+1])
        lister[i] = numer/denom
        lister[i+1] = numer/denom  
    return int(lister[len(lister) - 1])   

def keyscram(keys):
    effkeylen = LCM([len(i) for i in keys])
    counter = 0
    for i in keys:
        keys[counter] = i * int((effkeylen/len(i)))
        counter +=1
    for i in range (0, len(keys) - 1):
        keys[i+1] = vigenere(keys[i], keys[i + 1])
    keys = keys[len(keys)-1]
    return keys
    

    
def chooser():
    gate = 0
    while gate == 0:
        gate = input("Would You Like To Encode or Decode a message?")
        print("")
        gate = gate.lower()
        if (gate != 'encode') == True and (gate != 'decode') == True and (gate != 1) == True and (gate != 2) == True:
            gate = 0
            continue
            
    if 'encode' in gate or gate == 1:
        message = input('What Should Be Encoded?\n Your Message:    ')
    elif 'decode' in gate or gate == 2:
        message = input('What Should Be Decoded?\n Your Message:    ')
    print("")
    
    numciphers = 0
    while numciphers == 0:
        numciphers = input('How many keys do you have?\n Number of Keys:    ')
        try:
            numciphers = int(numciphers)
        except:
            numciphers = 0
            print("Please Input a Positive Number")
        if numciphers < 0:
            numciphers = 0
            print("Please Input a Positive Number")
    print("")
    
    keys = []
    counter = 1
    for i in range(0,int(numciphers)):
        key = input('Please Input Key #' + str(counter) + '\n Your Key:    ')
        print("")
        counter += 1
        keys.append(key)
    keys = keyscram(keys)
    
    if 'encode' in gate or gate == 1:
        print('Encrypting Your Message...')
        output = vigenere(message, keys)
    elif 'decode' in gate or gate == 2:
        print('Decrypting Your Message...')
        output = descrambler(message, keys)

    return output
```
