```python
indic = 'GIN'
indic2 = "Female legislators, senior officials "
country = 'USA'

mask1 = data['IndicatorName'].str.contains(indic)
mask2 = data['CountryCode'] == country
mask3 = data['IndicatorName'].str.contains(indic2)
mask4 = data['Year'] > 2003

arms = data[mask4 & mask2 & mask1]

mask5 = data ['Year'].isin(arms['Year'])

lex = data[mask2 & mask3 & mask5]
```

```python
years = arms['Year'].values
values = arms['Value'].values

plt.bar(years, values)
plt.show()

years = lex['Year'].values
values = lex['Value'].values

plt.bar(years, values)
plt.show()
```

```python
%matplotlib inline
from matplotlib.pyplot import figure

fig, axis = plt.subplots()

axis.yaxis.grid(True)
axis.set_title('GINI Index vs Woman In High Ranking Positions', fontsize = 10)
axis.set_xlabel(arms['IndicatorName'].iloc[0], fontsize = 10)
axis.set_ylabel(lex['IndicatorName'].iloc[0],fontsize = 10)

X = arms['Value']
Y = lex['Value']

axis.scatter(X,Y)
plt.show()
```

```python
from matplotlib.pyplot import figure
def plotter(countrylist):
    %matplotlib inline
    indic = 'AG.LND.ARBL.Z'
    indic2 = "Fertility rate, total "
    mask1 = data['IndicatorCode'].str.contains(indic)
    mask3 = data['IndicatorName'].str.contains(indic2)
    mask4 = data['Year'] > 1960
    for countryname in countrylist:
        try:
            country = countryname
            mask2 = data['CountryName'] == country

            arblnd = data[mask1 & mask2]
            fertrate = data[mask3 & mask2 & mask4]

            fig, axis = plt.subplots()
            axis.yaxis.grid(True)

            axis.set_title('Arable Lands vs Fertility Rates in ' + country, fontsize = 9)
            axis.set_xlabel(arblnd['IndicatorName'].iloc[0], fontsize = 10)
            axis.set_ylabel(fertrate['IndicatorName'].iloc[0],fontsize = 10)

            X = arblnd['Value']
            Y = fertrate['Value']

            print(np.corrcoef(X,Y))

            axis.scatter(X,Y)
            plt.show()
        except:
            continue
```

```python
plotter(countrylist)
```
